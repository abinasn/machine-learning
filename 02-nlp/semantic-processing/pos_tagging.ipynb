```python
import nltk
from nltk import word_tokenize, pos_tag

sentence = "Apple launched the new MacBook with faster processor"

tokens = word_tokenize(sentence)
print("Tokens:", tokens)

pos_tags = pos_tag(tokens)
print("\nPOS Tags:", pos_tags)

# Extract NOUNS
nouns = [word for word, tag in pos_tags if tag.startswith('NN')]
print("\nNouns (Products/Things):", nouns)

# Extract VERBS (different variable name!)
verbs = [word for word, tag in pos_tags if tag.startswith('VB')]
print("\nVerbs (Actions):", verbs)

sentence = "The faster car won the race"
tokens = word_tokenize(sentence)
pos_tags = pos_tag(tokens)
print(pos_tags)

```

DT = Determiner (the, a, an)
NN = Noun, singular
VBZ = Verb, 3rd person singular present
RB = Adverb
IN = Preposition

```python
review = "I bought the iPhone yesterday and love the camera quality"

tokens = word_tokenize(review)
pos_tags = pos_tag(tokens)

print("Full POS Tags:")
print(pos_tags)

nouns = [word for word, tag in pos_tags if tag.startswith('NN')]
print("\nNouns (Products/Things):", nouns)
```
